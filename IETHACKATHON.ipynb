```python
import pytesseract
pytesseract.pytesseract.tesseract_cmd=r'C:\Program Files\Tesseract-OCR\tesseract.exe'
import keras_ocr
import easyocr
import matplotlib.pyplot as plt
import cv2
get_ipython().run_line_magic('matplotlib', 'inline')
```

```python
image_path = [
    "https://i.redd.it/7wpmbsckae371.jpg",
     r"C:\Users\trc\Desktop\22.jpg",
    "https://raw.githubusercontent.com/sanskar-hasija/ocr-comparision/main/test_images/image3.png",
    "http://test.ultralinx.co/wp-content/uploads/2020/03/1_beautiful-handwriting-lettering-calligraphy-58-572b35d8d1992__700jpg.jpg",
]
images = [ keras_ocr.tools.read(i) for i in image_path]
```

```python
fig = plt.figure(figsize=(16,10))
rows = 2
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(images[0])
plt.axis('off')
plt.title("First Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(images[1])
plt.axis('off')
plt.title("Second Image")

fig.add_subplot(rows, columns, 3)
plt.imshow(images[2])
plt.axis('off')
plt.title("Third Image")

fig.add_subplot(rows, columns, 4)
plt.imshow(images[3])
plt.axis('off')
plt.title("Fourth Image");
```

```python
pipline = keras_ocr.pipeline.Pipeline() #Creting a pipline 
kerasocr_preds = pipline.recognize(images)
```

```python
fig,axs = plt.subplots(nrows = 4 , figsize = (30,30))
for ax , image,  prediction in zip(axs , images , kerasocr_preds):
    keras_ocr.tools.drawAnnotations(image, prediction, ax)
```
